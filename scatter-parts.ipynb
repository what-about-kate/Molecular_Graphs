```python
%matplotlib inline

import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
name = 'bzr'
x = []
with open("bzr-Wiener_indices.csv") as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        arr = [float(i) for i in row]
        x.append(np.sqrt(arr[0]))
        
#print (x)
print (len(x))
```

```python
y = []
with open("bzr-Randich_indices.csv") as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        arr = [float(i) for i in row]
        y.append(arr[0])
        
#print (y)
print (len(y))
```

```python
xx = []
yy = []

for i in range(len(x)) :
    if x[i] >= 12 and x[i] <= 50 :
        xx.append(x[i])
        yy.append(y[i])
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y, c='blue')

ax.set_facecolor('white')
ax.set_title(name, fontsize=20, color='black')

ax.set_xlabel("sqrt (Wiener index)", fontsize=15, color='black', labelpad=10)
ax.set_ylabel("Randich index", fontsize=15, color='black', labelpad=10)

fig.set_figwidth(20)
fig.set_figheight(10)

plt.show()

#fig.savefig(name)
```

```python
fig, ax = plt.subplots()

ax.scatter(xx, yy, c='green')

ax.set_facecolor('white')
ax.set_title(name, fontsize=20, color='black')

ax.set_xlabel("sqrt (Wiener index)", fontsize=15, color='black', labelpad=10)
ax.set_ylabel("Randich index", fontsize=15, color='black', labelpad=10)

fig.set_figwidth(20)
fig.set_figheight(10)

plt.show()

str = name + '_1'
fig.savefig(str)
```
